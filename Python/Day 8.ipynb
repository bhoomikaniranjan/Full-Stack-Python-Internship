```python
'''
Question:
Write a function describe_person that takes a variable number of keyword arguments representing a person's attributes (e.g., name, age, city, etc.) and prints them in a readable format. The function should accept arbitrary keyword arguments and print each key-value pair in a formatted manner.

Input:
	* A variable number of keyword arguments (using the **kwargs syntax), where each keyword represents an attribute of a person, and its value represents the corresponding attribute value.
	* Use kwargs in your implementation.

Output:
	* The function should print each attribute in the format key: value.

Constraints:
	* The number of keyword arguments can vary (i.e., there can be any number of attributes).
	* Input kwargs must be of datatype dictionary.

'''

def describe_person(**kwargs):
    # WRITE YOUR CODE HERE...
    for key, value in kwargs.items():
        print(f"{key}:{value}")

kwargs = {'name': 'sujay', 'age': 20, 'Occupation': 'Engineer'}
describe_person(**kwargs)
```

```python
'''
Write a function max_of_all that takes a variable number of integer arguments and returns the maximum value among them. The function should be able to handle any number of arguments, and return the largest value passed to it.

Input:
	* A variable number of integer arguments (using the *args syntax), representing a list of integers.
	* Use args in your implementation.

Output:
	* The maximum integer value from the provided arguments.
	* If the list is empty, return "Empty List" if it's empty.

Constraints:
	* The integers passed should be valid integers, if passed remove non integers and consider the max among the others.
	* Input args must be of datatype list. The elements in the list must be integers.

'''

def max_of_all(*args):
    # WRITE YOUR CODE HERE...
    valid = [arg for arg in args if isinstance(arg,int)]
    if not valid:
        return "Empty List"
    return max(valid)

args = [-1, -2,-3,5]
print(max_of_all(*args))
```

```python
'''
Write a decorator add_10 that adds 10 to the result of the function it decorates. Then, apply this decorator to a function multiply_by_5(x) that multiplies x by 5.

Input:
	* An integer x is passed to the function multiply_by_5(x).
	* Use x in your implementation.

Output:
	* The function multiply_by_5(x) multiplies x by 5, and then the add_10 decorator adds 10 to the result. The final output is the result of this operation.

Constraints:
	* Input x should be integer within the range -10**3 to 10**3.
'''

def add_10(func):
    # WRITE YOUR CODE HERE...
    def wrapper(x):
        result = func(x)
        return result + 10
    return wrapper

@add_10
def multiply_by_5(x):
    # WRITE YOUR CODE HERE...
    return x * 5

x = -15
print(multiply_by_5(x))
```

```python
'''
Write a function merge_dicts that takes any number of dictionaries as keyword arguments and returns a single dictionary that merges all the input dictionaries. In case of key conflicts (where the same key appears in multiple dictionaries), later dictionaries should overwrite earlier ones.

Input:
	* A variable number of dictionaries passed as keyword arguments (using **kwargs), where each dictionary represents a set of key-value pairs.
	* Use kwargs in your implementation.

Output:
	* A single dictionary that contains all key-value pairs from the input dictionaries. In case of conflicting keys, the value from the last dictionary (in the order they were passed) will be used.

Constraints:
	* The number of dictionaries can vary (i.e., there can be any number of dictionaries passed).
	* kwargs is a dictionary where each key maps to a nested dictionary.
'''

def merge_dicts(**kwargs):
    # WRITE YOUR CODE HERE...
    d = {}
    for k,v in kwargs.items():
        d.update(v)
    return d

kwargs = {"a": {'name': 'John'}, "b": {'age': 30}, 'c': {'occupation': 'Engineer'}, 'd': {'age': 35}}
print(merge_dicts(**kwargs))
```

```python
'''
Write a function sum_all that takes a variable number of integer arguments and returns their sum. The function should accept an arbitrary number of arguments and the sum of all these numbers should be computed and returned.

Input:
	* A variable number of integer arguments (using the *args syntax).
	*  Use args in your implementation.

Output:
	* The sum of all the provided integers.

Constraints:
	* The integers passed should be valid integers within the range -10**3 to 10**3.
	* Input args should be list of numeric values.
'''

def sum_all(*args):
    # WRITE YOUR CODE HERE...
    s = 0
    for i in args:
        s += i
    return s

args = [4, 5, 9, 2]
print(sum_all(*args))
```

```python
'''
Write a decorator cache that stores the results of function calls and returns the cached result if the function is called with the same arguments. Apply this decorator to a function fibonacci(n) that computes the nth Fibonacci number.

Input:
	* An integer n representing the position in the Fibonacci sequence for which the value is to be computed.
	* Use n in your implementation.

Output:
	* The nth Fibonacci number, computed using recursion, with caching to avoid redundant calculations.
	* Print "Invalid Number" if n not in range 0 and 100 (inclusive).

Constraints:
	* Input n must be of type integer.

'''

def cache(func):
    # WRITE YOUR CODE HERE...
    cache_dict = {}
    def wrapper(n):
        if n in cache_dict:
            return cache_dict[n]
        else:
            result = func(n)
            cache_dict[n] = result
            return result
    return wrapper

@cache

def fibonacci(n):
    # WRITE YOUR CODE HERE...
    if not (0 <= n <= 100):
        return "Invalid Number"
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibonacci(n-1) + fibonacci(n-2)

n = 15
print(fibonacci(n))
```

```python
'''
Write a function create_greeting that takes a variable number of positional arguments for names and a variable number of keyword arguments for additional information like greeting and punctuation. The function should return a greeting message that combines the names with the greeting and punctuation provided.

Input:
	* A variable number of positional arguments (*args) representing names.
	* A variable number of keyword arguments (**kwargs) representing additional information:
	* greeting: The greeting to use (default: "Hello").
	* punctuation: The punctuation to use at the end (default: "!").
	* Use args and kwargs in your implementation.

Output:
	* A formatted greeting message based on the input names, greeting, and punctuation.
	* The greeting and punctuation are optional keyword arguments. If not provided, defaults will be used (greeting="Hello", punctuation="!").
	* The names should be concatenated with the word "and" between the last two names (if more than one name is provided).

Constraints:
	* The number of names can vary.
	* Input kwargs must be of datatype dictionary with 2 keys punctuation and greeting.
	* Input args must be of datatype list.
'''

def create_greeting(*args, **kwargs):
    # WRITE YOUR CODE HERE...
    greeting = kwargs.get("greeting","Hello")
    punctuation = kwargs.get("punctuation","!")
    if len(args) > 1:
        names = ",".join(args[:-1]) + " and " +args[-1]
    elif len(args) == 1:
        names = args[0]
    else:
        names = ''
    return f"{greeting}, {names}{punctuation}"

args = ["Tom", "Jerry"]
kwargs = {"greeting": "Hello", "punctuation": "!"}
print(create_greeting(*args, **kwargs))
```

```python
'''
Write a decorator only_positive that only allows a function to execute if its argument is positive. Apply this decorator to a function square(x) that returns x squared.

Input:
	* A single integer x.
	* Use x in your implementation.

Output:
	* If x is positive, the function returns the square of x.
	* If x is not positive (i.e., x <= 0), the function returns the error message "Error: Argument must be positive".

Constraints:
	* Input x should be integer with in the range -10**3 to 10**3.
'''

def only_positive(func):
    # WRITE YOUR CODE HERE...
    def wrapper(x):
        if x > 0:
            return func(x)
        else:
            return "Error: Argument must be positive"
    return wrapper

@only_positive

def square(x):
    # WRITE YOUR CODE HERE...
    return x * x

x = 5
print(square(x))
```

```python
'''
Write a decorator memoize that caches the results of the function it decorates to avoid redundant calculations. Apply this decorator to a function factorial(n) that computes the factorial of n.

Input:
	* An integer n for which the factorial is to be calculated.
	* Use n in your implementation.

Output:
	* The factorial of n, computed using a recursive approach, with caching to optimize repeated calculations.
	* Print "Invalid Number" if n not in range 0<= n <= 100.

Constraints:
	* Input n must be integer.
'''

def memoize(func):
    # WRITE YOUR CODE HERE...
    cache = {}
    def wrapper(n):
        if n in cache:
            return cache[n]
        result = func(n)
        cache[n] = result
        return result
    return wrapper

@memoize

def factorial(n):
    # WRITE YOUR CODE HERE...
    if not isinstance(n, int) or n < 0 or n > 100:
        return "Invalid Number"
    if n == 0 or n == 1:
        return 1
    return n * factorial(n - 1)

n = 25
print(factorial(n))
```

```python
'''
Question:
Write a decorator named decorator that prints a statement and executes the function it decorates. Then, apply this decorator to a function calculate_sum(n) that calculates the sum of numbers from 1 to n (inclusive).

Input:
	* An integer n, representing the upper limit (inclusive) up to which the sum is to be calculated.
	* Use n in your implementation.

Output:
	* The function calculate_sum(n) calculates the sum of numbers from 1 to n (inclusive) and returns the result. Additionally, the decorator prints a statement.
	* Print "Invalid Number" if not in range 1<= n <= 10**6.

Constraints:
	* Input n should be integer.
'''

def decorator(func):
    # WRITE YOUR CODE HERE...
    def wrapper(n):
        print("Inside the decorator")
        return func(n)
    return wrapper

@decorator

def calculate_sum(n):
    # WRITE YOUR CODE HERE...
    if not (1 <= n <= 10 ** 6):
        return "Invalid Number"
    return sum(range(1, n + 1))

n = -1343
print(calculate_sum(n))
```
