```python
'''
You are tasked with writing a Python script that finds the maximum number in a list using a "for" loop.

Input:
	* A list of integers or floats numbers from which to find the maximum number.
        * Use input numbers in your implementation

Output:
	* Print the maximum number found in the list.

Constraints:
	* The list must contain at least one element.
	* All elements in the list should be numbers (integers or floats).

'''
def find_max_number(numbers):
    # WRITE YOUR CODE HERE...
    if not numbers:
        max_number = 0
    else:
        max_number = numbers [0]
        for i in numbers:
            if i > max_number:
                max_number = i
    print(max_number)
numbers = [11145, 11456, 10101, 11001]
find_max_number(numbers)
```

```python
'''
You are tasked with writing a Python function that counts how many times a specific number appears in a list of numbers using a "for" loop.

Input:
	* A list of integers or floats numbers_list in which to count occurrences.
	* A number (integer or float) number that you want to count in the list.
        * Use numbers_list and number in your implementation.

Output:
	* Print the count of how many times the specified number appears in the list.

Constraints:
	* The list may contain any number of elements (including zero elements).
	* The number can be any integer or float.
'''

def occurrence_of_numbers(numbers_list, number):
    # WRITE YOUR CODE HERE...
    count = 0
    for num in numbers_list:
        if num == number:
            count += 1
    print(count)

numbers_list = [1,2,3,4,5,6,7,1,2]
number = 4
occurrence_of_numbers(numbers_list, number)
```

```python
'''
Write a python script count_digits that counts the number of digits in a given number using a "for" loop. The script should return the total number of digits in the number.

Input:
	* A single integer n (0 <= n <= 100000), representing the number to count the digits in it.
	* Use input n in your implementation.

Output:
	* The script should return an integer, which represents the number of digits in the number n.
	* Print "Invalid Number", if (n<0 or n>100000).

Constraints:
	* The input n should be a integer.
'''

def count(n):
    if n < 0 or n > 100000:
        print("Invalid Number")
        return
    count = 0
    for digit in str(n):
        count += 1
    return count

n = 123457
result = count(n)
if result is not None:
    print(result)
```

```python
'''
You are tasked with writing a Python program using "while" loop that calculates and prints the average temperature for a week based on user inputs. The program should prompt the user to input daily temperatures for 7 days.

Input:
	* A list named temp_list containing exactly 7 daily temperatures.
	* Use temp_list variable in your implementation.

Output:
	* Print the average temperature for the week, rounded to two decimal places.
	* Print "Invalid" If length of temp_list is not equal to 7.

Constraints:
	* The input temp_list must be  a list of numbers.
'''

def avg_temperature(temp_list):
    # WRITE YOUR CODE HERE...
    if len(temp_list) != 7:
        print("Invalid")
        return
    total_temp = 0
    i = 0
    while i < len(temp_list):
        total_temp += temp_list[i]
        i += 1
    avg = total_temp / len(temp_list)
    print(f"Average temperature for the week : {avg:.2f}")

temp_list = [20.5,22.0,21.5,19.0,18.5,26.5,21.0]
avg_temperature(temp_list)
```

```python
'''
Write a python script sum_of_numbers that calculates the sum of all numbers from 1 to N using a "for" loop. The function should return the total sum of numbers starting from 1 up to N (inclusive).

Input:
	* A single integer n (0<= n <=1000), representing the upper limit of the sum (inclusive).
	* You must use n in your implementation.

Output:
	* The python script should return the sum of all numbers from 1 to n.
	* Print "Invalid Number" If (n<0 or n>1000).

Constraints:
	* The input n must be a integer.
'''
def sum_of_num(n):
    if n < 0 or n > 1000:
        print("Invalid Number")
        return
    total_sum = 0
    for i in range (1,n + 1):
        total_sum += i
    return total_sum

n = 1293
result = sum_of_num(n)
if result is not None:
    print(result)
```

```python
'''
You are tasked with writing a Python program that prints the multiplication table for a given number using a "for" loop.

Input:
	* An integer number for which the multiplication table will be generated.
        * Use input number in your implementation.

Output:
	* Print the multiplication table for the given number from 1 to 10.
       * "Invalid input." if number is less than or equal to 0.

Constraints:
	* The number must be a integer.
'''

def generate_multiplication_table(number):
    # WRITE YOUR CODE HERE...
    if number <= 0:
        print("Invalid input.")
    else:
        for i in range (1,11):
            print(f"{number} x {i} = {number * i}")
number = 10
generate_multiplication_table(number)
```
