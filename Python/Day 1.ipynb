```python
'''
Question:
You are tasked with writing a Python script  using 'If' statement that evaluates a student's grade based on their score.

Input:
	* An integer score representing the student's score.
        * You must use the variable "score" in your implementation.

Output:
	* Print "A" if the score is 90 or above.
	* Print "B" if the score is between 80 and 89 (inclusive).
	* Print "C" if the score is between 70 and 79 (inclusive).
	* Print "D" for any score below 70.
        * Print "Invalid score" if score is less than 0 or greater than 100.

Constraints:
        * Input score must be an integer.
'''

score = int(input("Enter the score: "))
if score < 0 or score > 100:
    print("Invalid score")
elif score >= 90:
    print("A")
elif score >= 80:
    print("B")
elif score >= 70:
    print("C")
else:
    print("D")
```

```python
'''
Write a Python script to classify a triangle based on the lengths of its three sides. The triangle should be classified as "Equilateral", "Isosceles", or "Scalene" using nested if-else conditions.

Input:
	* Three integer values a, b, and c representing the lengths of the triangle's sides.
	* You must use the variables a, b, and c in your implementation.

Output:
	* A string indicating the type of triangle: "Equilateral" if all three sides are equal, "Isosceles" if exactly two sides are equal, and "Scalene" if all sides are different.
        * Print "Invalid side lengths"  if inputs are not in range (1 <= a, b and c<= 1000)
        * If the inputs are in range and the triangle cannot be formed then print "Invalid triangle".

Constraints:
	* Each side length a, b, and c must be integer.
'''

def classify_triangle(a,b,c):
    if not (1 <= a <= 1000 and 1 <= b <= 1000 and 1 <= c<= 1000):
        print("Invalid side lengths")
    else:
        if a + b > c and b + c > a and c + a > b:
            if a == b == c:
                print("Equilateral")
            elif a == b or b == c or a == c:
                print("Isosceles")
            else:
                print("Scalene")
        else:
            print("Invalid triangles")

a = int(input())
b = int(input())
c = int(input())
classify_triangle(a,b,c)
```

```python
'''
Write a Python function to determine the quadrant of a point (x, y) in the Cartesian plane.

Input:
	* Two integer values x and y representing the coordinates of a point on the Cartesian plane.
	* You must use the variables x and y in your implementation.

Output:
	* A string indicating the quadrant of the point:
	* "Quadrant 1" if (x, y) is in the first quadrant (both x and y are positive).
	* "Quadrant 2" if (x, y) is in the second quadrant (x is negative, y is positive).
	* "Quadrant 3" if (x, y) is in the third quadrant (both x and y are negative).
	* "Quadrant 4" if (x, y) is in the fourth quadrant (x is positive, y is negative).
	* "Origin" if both x and y are zero.
        * "Invalid Input" if x and y are not in range (-10,000 and 10,000) exclusive.

Constraints:
	* "x" and "y" should be a integer.
  '''
def determine_quadrant(x, y):
    # WRITE YOUR CODE HERE...
    if -10000 < x < 10000 and -10000 < y < 10000:
        if x>0 and y>0:
            return "Quadrant 1"
        elif x<0 and y >0:
            return "Quadrant 2"
        elif x <0 and y <0:
            return "Quadrant 3"
        elif x>0 and y<0:
            return "Quadrant 4"
        elif x==0 and y==0:
            return "Orgin"
    else:
        return "Invalid Input"
x = 3
y = 45425654
print(determine_quadrant(x,y))
```

```python
'''
You are tasked with writing a Python function to calculate the final price after applying a discount based on the total purchase amount.

Input:
	* A float price representing the total purchase amount (0 < price ≤ 10,000).
	* You must use the variable "price" in your implementation.

Output:
	* Print a string representing the final price after applying the discount, rounded to two decimal places, formatted as "$X.XX" where X.XX is the final price.
        * Discount Rules:
	    If the price is greater than or equal to $100, apply a 10% discount.
	    If the price is less than $100, apply a 5% discount.
	    If the price is -ve or greater than $10000, print a "Invalid Price".

Constraint:
        * The input price should be a number.
'''
price = float (input("Enter a price :"))
if 0 < price <= 10000:
    if price >= 100:
        discount = 0.10
    else:
        discount = 0.05
    final = price - (price * discount)
    print(f"${final:.2f}")
else:
    print("Invalid Price")
```

```python
'''
Write a Python program to determine if a number is divisible by 3, 5, or both using if statements.

Input:
	* An integer num representing the number to check for divisibility.
	* You must use the variable num in your implementation.

Output:
	* If the number is divisible by both 3 and 5, print "Divisible by both".
	* If the number is divisible by only 3, print "Divisible by 3".
	* If the number is divisible by only 5, print "Divisible by 5".
	* If the number is not divisible by either 3 or 5, print "Not divisible by 3 or 5".
        * 0 ≤ num ≤ 1000, If input range fails print "Invalid Number".

Constraints:
	* Input num must be a integer.

'''

num = int(input("Enter number"))
if 0 <= num <= 1000:
    if num %3 == 0 and num %5 == 0:
        print("Divisible by both")
    elif num %3 == 0:
        print("Divisible by 3")
    elif num %5 == 0:
        print("Divisible by 5")
    else:
        print("Not divisible by 3 or 5")
else:
    print("Invalid Number")
```
