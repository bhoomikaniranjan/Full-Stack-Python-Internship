```python
'''
Write a Python program that uses a list comprehension to create a list of even numbers from 1 to a given limit. The program should print the list of even numbers.

Input:
	* A positive integer limit (2 <= limit <= 100) representing the upper bound (exclusive) for the range of numbers.
	* Use limit in your implementation

Output:
	* A list of even numbers from 1 to limit (excluding limit).
	* Print "Invalid Limit". If limit not in range(2<= limit <= 100).

Constraints:
	* Input limit must be a integer.
'''

def even_number(limit):
    # WRITE YOUR CODE HERE...
    if isinstance(limit,int) and 2 <= limit <= 100:
        even_number = [num for num in range(2,limit) if num % 2 == 0]
        print(even_number)
    else:
        print("Invalid Limit")

limit = 100
even_number(limit)
```

```python
'''
Question:
Write a Python program that uses a list comprehension to create a list of the squares of numbers from 1 to a given limit (exclusive). The program should print the list of squared numbers.

Input:
	* A positive integer limit (2 <= limit <= 100) representing the upper bound (exclusive) for the range of numbers.
	* Use limit in your implementation.

Output:
	* A list of the squares of numbers from 1 to (limit - 1).
	* Print "Invalid Limit". If limit not in range(2<= limit <= 100).

Constraints:
	* Input limit must be a integer.
'''

def square_number(limit):
    # WRITE YOUR CODE HERE...
    if limit < 1 or limit > 100:
        print("Invalid Limit")
    else:
        l = []
        for i in range (1,limit):
            l = l + [i ** 2]
        print(l)

limit = 50
square_number(limit)
```

```python
'''
Write a Python function that takes a string as input and iterates through its characters. Skip any character that is a digit using a "continue" statement, and print only the non-digit characters. If no non-digit characters are found in the string, output "No non-digit characters found".

Input:
	* A single string of datatype string.
	* Use string in your implementation

Output:
	* For each non-digit character in the string, print the character.
	* If no non-digit characters are found in the string, print "No non-digit characters found.".
	* Print "Error: Input string must contain 1-100 characters." if not in range (1 <= len(string) <= 100).

Constraints:
	* Input string must be datatype of string.

'''

def print_non_digit_characters(string):
    # WRITE YOUR CODE HERE...
    if len(string) < 1 or len(string) > 100:
        print("Error: Input string must contain 1-100 characters.")
    else:
        count = 0
        for i in string:
            if i.isdigit():
                count += 1
                continue
            print(i)
        if count == len(string):
            print("No non-digit characters found.")

string = "Karthi19"
print_non_digit_characters(string)
```

```python
'''
Write a Python function that takes a list of words as input and iterates through them. Skip any word that is shorter than 4 letters using a "continue" statement, and print only the remaining words. If no words meet the length criteria, print "No words processed".

Input:
	* A list words containing strings, where each string is a word.
	* Use words in your implementation.

Output:
	* For each word that is 4 or more letters, print the word.
	* If all words in the list are shorter than 4 letters, print "No words processed".
	* Print "Error: Each word must have 1-20 characters and the list must contain 1-100 words." If length of any word in the list is less than 1 or greater than 20 or If length of list is not in range (1 <= len(words) <= 100).

Constraints:
	* Input words must be list of strings.
'''

def print_long_words(words):
    # WRITE YOUR CODE HERE...
    if (len(words) < 1 or len(words) > 100) or any (len(h) < 1 or len(h) > 20 for h in words):
        print("Error: Each word must have 1-20 characters and the list must contain 1-100 words.")
    else:
        count = 0
        for i in words:
            if len(i) < 4:
                count += 1
                continue
            print(i)
        if count == len(words):
            print("No words processed")


words = ["cat", "apple", "dog", "banana", "fig","Karthi"]
print_long_words(words)
```

```python
'''
Write a Python function that takes a list of words as input and iterates through them. Skip any word that contains the letter 'e' using a "continue" statement, and print only the remaining words. If no words are printed, output "No words processed".

Input:
	* A list words containing strings, where each string is a word.
	* Use words in your implementation.

Output:
	* For each word that does not contain the letter 'e', print the word.
	* If all words in the list contain "e" then print "No words processed".
	* Print "Error: Each word must have 1-20 characters and the list must contain 1-100 words." If length of any word in the list is less than 1 or greater than 20 or If length of list is not in range (1 <= len(words) <= 100).

Constraints:
	* Input words must be list of strings.
'''

def print_words_without_e(words):
    # WRITE YOUR CODE HERE...
    if (len(words) < 1 or len(words) > 100) or any (len(h) < 1 or len(h) > 20 for h in words):
        print("Error: Each word must have 1-20 characters and the list must contain 1-100 words.")
    else:
        count = 0
        for i in words:
            if any(b in 'e' for b in i):
                count += 1
                continue
            print(i)
        if len(words) == count:
            print("No words processed")

words =  ["apple", "banana", "fig", "grape", "plum"]
print_words_without_e(words)
```

```python
'''
Write a script that skips uppercase letters Write a Python function that takes a string as input and iterates through its characters. Skip any character that is an uppercase letter using a "continue" statement, and print only the lowercase letters. If no lowercase letters are found in the string, output "No lowercase letters found".

Input:
	* A single string of datatype string.
	* Use string in your implementation

Output:
	* For each lowercase letter in the string, print the letter.
        * If no lowercase characters are found in the string, print "No lowercase letters found".
	* Print "Error: Input string must contain 1-100 characters." if not in range (1 <= len(string) <= 100).

Constraints:
	* Input string must be datatype of string.
'''

def print_lowercase_characters(string):
    # WRITE YOUR CODE HERE...
    if len(string) < 1 or len(string) > 100:
        print("Error: Input string must contain 1-100 characters.")
    else:
        count = 0
        for i in string:
            if i.isupper():
                count += 1
                continue
            elif i.islower():
                print(i)
            else:
                count += 1
        if count == len(string):
            print("No lowercase letters found")

string = "pythonCode"
print_lowercase_characters(string)
```
