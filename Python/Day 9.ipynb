```python
'''
Write a function "is_even(x)" that returns True if the number x is even, and False otherwise. Then, implement a function "filter_even(numbers)" that uses the "filter" function to apply is_even to a dynamically passed list of numbers, and returns a list of all even numbers from the list.

Input:
	* A list of integers numbers, where each integer can be positive, negative, or zero.
	* Use numbers in your implementation.

Output:
	* A list containing only the even numbers from the numbers list.
	* Result must be "List cannot be empty." if length of the list is 0.
	* Result must be "Invalid Input." if length of the list is greater than 1000 or if any value in the list is greater than 1000.

Constraints:
	* Input numbers is list of integers.
'''

def is_even(x):
    # WRITE YOUR CODE HERE...
    return x % 2 == 0

def filter_even(numbers):
    # WRITE YOUR CODE HERE...
    if len(numbers) == 0:
        return "List cannot be empty."
    if len(numbers) > 1000 or any(abs(num) > 1000 for num in numbers):
        return "Invalid Input."
    even_numbers = filter(is_even, numbers)
    return list(even_numbers)

numbers = []
print(filter_even(numbers))
```

```python
'''
Write a Python function "halve_greater_than_fifty" using "filter" function to filter numbers greater than 50 from a list and return each of them halved.

Input:
	* A list of integers numbers where each element is an integer.
	* Use numbers in your implementation.

Output:
	* A list of floats, where each element is a number from numbers that is greater than 50, divided by 2.
	* Result must be "Invalid input: List exceeds maximum length of 1,000 integers." if length of the list is greater than 1000.
	* Result must be "Invalid input: Integers must be between -10,000 and 10,000." if any value in the list is greater than 10000 or less than -10000.

Constraints:
	* Input numbers is list of integers.
'''

def halve_greater_than_fifty(numbers):
    # WRITE YOUR CODE HERE...
    if len(numbers) > 1000:
        return "Invalid input: List exceeds maximum length of 1,000 integers."
    if any(num < -10000 or num > 10000 for num in numbers):
        return "Invalid input: Integers must be between -10,000 and 10,000."
    filters = map(lambda x: x / 2.0, filter(lambda x: x > 50, numbers))
    return list(filters)

numbers = [60, 30, 100, 45, 80]
print(halve_greater_than_fifty(numbers))
```

```python
'''
Write a Python function "short_strings" using "filter" function to filter strings shorter than 3 characters from a list and return them in a new list.

Input:
	* A list of strings strings where each element is a string.
	* Use strings in your implementation.

Output:
	* A list of strings containing only the elements from strings that are shorter than 3 characters.
	* Result must be "Invalid Input." if length of the list is greater than 1000 or if length of any value in the list is greater than 100.
	* Result must be 0 if length of list is 0.

Constraints:
	* Input strings is list of datatype string.
	* Each string in the input list strings will have a length between 1 and 100 characters. The list strings will contain at most 1,000 strings.
	* Strings can include uppercase and lowercase letters, digits.

'''

def short_strings(strings):
    # WRITE YOUR CODE HERE...
    if len(strings) > 1000:
        return "Invalid Input."
    if any(len(s) > 100 for s in strings):
        return "Invalid Input."
    if len(strings) == 0:
        return 0
    short = filter(lambda s: len(s) < 3, strings)
    return list(short)

strings =  ["a", "to", "be", "cat", "do"]
print(short_strings(strings))
```

```python
'''
Write a function "is_long_string(s)" that returns True if the length of the string s is greater than 5, and False otherwise. Then, implement a function "filter_long_strings(strings)" that uses the "filter" function to apply is_long_string to a dynamically passed list of strings, and returns a list of all strings with a length greater than 5.

Input:
	* A list of strings strings, where each string can contain alphabetic characters, numbers, and other symbols.
	* Use strings in your implementation.

Output:
	* A list containing only the strings from strings whose length is greater than 5.
	* Result must be "Invalid Input." if length of the list is greater than 1000 or if length of any value in the list is greater than 100.

Constraints:
	* Input strings is list of datatype string.
'''

def is_long_string(s):
    # WRITE YOUR CODE HERE...
    return len(s) > 5

def filter_long_strings(strings):
    # WRITE YOUR CODE HERE...
    if len(strings) > 1000 or any(len(s) > 100 for s in strings):
        return "Invalid Input."
    return list(filter(is_long_string,strings))

strings = ["apple", "banana", "cherry", "kiwi", "pear"]
print(filter_long_strings(strings))
```

```python
'''
Write a Python function "add_ten_to_negatives" using "filter" function to filter out negative numbers from a list and add 10 to each of them.

Input:
	* A list of integers numbers where each element is an integer.
	* Use numbers in your implementation.

Output:
	* A list of integers, where each element is a negative number from numbers with 10 added to it.
	* Result must be "Invalid input: List exceeds maximum length of 1,000 integers." if length of the list is greater than 1000.
	* Result must be "Invalid input: Integers must be between -10,000 and 10,000." if any value in the list is greater than 10000 or less than -10000.

Constraints:
	* Input numbers is list of integers.
'''

def add_ten_to_negatives(numbers):
    # WRITE YOUR CODE HERE...
    if len(numbers) > 1000:
        return "Invalid input: List exceeds maximum length of 1,000 integers."
    for num in numbers:
        if num < -10000 or num > 10000:
            return "Invalid input: Integers must be between -10,000 and 10,000."
    neg_numbers = filter(lambda x: x < 0, numbers)
    result = [x + 10 for x in neg_numbers]
    return result

numbers = [10, -20, -1, -100, 50]
print(add_ten_to_negatives(numbers))
```

```python
'''
You are given a list of integers. Write a function "cube_numbers" that uses "map" function to return a new list where each element is the cube of the corresponding element from the input list.

Input:
	* A list of integers numbers where each element represents a number to be cubed (1 ≤ len(numbers) ≤ 1000).
	* Use numbers in your implementation.

Output:
	* A list of integers where each element is the cube of the corresponding element from the input list.
	* Result must be "The length of the input list must be between 1 and 1000." if length of the list is not in range 1 and 1000 (inclusive).

Constraints:
	* Input numbers is list of integers.
	* The list numbers will contain integers or floats within the range -10**3 to 10**3.
'''

def cube_numbers(numbers):
    # WRITE YOUR CODE HERE...
    if len(numbers) < 1 or len(numbers) > 1000:
        return "The length of the input list must be between 1 and 1000."
    cubed_numbers = map(lambda x: x ** 3, numbers)
    return list(cubed_numbers)

numbers = [1, 2, 3]
print(cube_numbers(numbers))
```

```python
'''
You are given a list of radii. Write a function "calculate_areas" that uses "map" function to calculate the area of a circle (using function "area_of_circle") for each radius in the list and returns a list of areas. Use the formula :
Area = round(π × radius ^ 2) where π is the mathematical constant approximately equal to 3.141

Input:
	* A list of integers or floats radius where each element represents the radius of a circle (1 ≤ len(radius) ≤ 10^4, 0 ≤ radius[i] ≤ 10^3).
	* Use radius in your implementation.
	* Note: Take pi value as 3.141

Output:
	* A list of floats where each element is the rounded area of a circle with the corresponding radius from the input list.

Constraints:
	* Input radius is list of integers. The list radius contains up to 10^4 elements. Each radius value is between 0 and 10^3.
'''

def area_of_circle(radius):
    # WRITE YOUR CODE HERE...
    pi = 3.141
    area = pi * (radius ** 2)
    return round(area)

def calculate_areas(radius):
    # WRITE YOUR CODE HERE...
    return list(map(area_of_circle,radius))

radius = [1,2,3]
print(calculate_areas(radius))
```

```python
'''
Write a function is_positive(x) that returns True if the number x is positive, and False otherwise. Then, implement a function "filter_positive(numbers)" that uses the "filter" function to apply "is_positive" to a dynamically passed list of numbers, and returns a list of all positive numbers from the list.

Input:
	* A list of integers numbers, where each integer can be positive, negative, or zero.
	* Use numbers in your implementation.

Output:
	* A list containing only the positive numbers from the numbers list.
	* Result must be "List cannot be empty." if length of the list is 0.
	* Result must be "Invalid Input." if length of the list is greater than 1000 or if any value in the list is greater than 1000.

Constraints:
	* Input numbers is list of integers.
'''

def is_positive(x):
    # WRITE YOUR CODE HERE...
    return x > 0

def filter_positive(numbers):
    # WRITE YOUR CODE HERE...
    if len(numbers) == 0:
        return "List cannot be empty."

    if len(numbers) > 1000 or any(abs(num) > 1000 for num in numbers):
        return "Invalid Input."
    positive_numbers = list(filter(is_positive, numbers))
    return positive_numbers

numbers = [11, 24, -65, 87, -12, 67, -84, -45, 62]
print(filter_positive(numbers))
```

```python
'''
Write a Python function "lengths_of_long_strings" using "filter" function to filter strings that are longer than 5 characters and return their lengths.

Input:
	* A list of strings strings where each element is a string.
	* You must use the variable strings in your implementation.

Output:
	* A list of integers representing the lengths of each string in strings that is longer than 5 characters.
	* Result must be 0 if length of the list is 0.
	* Result must be "Invalid Input." if length of the list is greater than 1000 or if length of any value in the list is greater than 100.

Constraints:
	* Input strings is list of datatype string.
'''

def lengths_of_long_strings(strings):
    # WRITE YOUR CODE HERE...
    if len(strings) > 1000 or any(len(s) > 100 for s in strings):
        return "Invalid Input."
    if not strings:
        return 0
    long_strings = filter(lambda s:len(s) > 5, strings)
    return [len(s) for s in long_strings]

strings = ["short", "tiny", "lengthy", "grandiose"]
print(lengths_of_long_strings(strings))
```

```python
'''
You are given two lists of integers, list1 and list2, of the same length. Write a function "add_lists" that uses "map" function to return a new list where each element is the sum of the corresponding elements from list1 and list2.

Input:
	* Two lists of integers, list1 and list2, of equal length (1 ≤ len(list1), len(list2) ≤ 10^3).
	* Use list1 and list2 in your implementation.

Output:
	* A list of integers where each element is the sum of the corresponding elements from list1 and list2.
	* Result must be "Both lists must have the same length." if length of both the list is not same.
	* Result must be "The length of the lists must be between 1 and 1000." if length of anyone of the list is not in range between 1 and 1000 (inclusive). (Hint: This condition should be checked if length of list1 and list2 is same)

Constraints:
	* Both list1 and list2 are list of integers. The elements of the list must be in range -10**3 to 10**3.

'''

def add_lists(list1, list2):
    # WRITE YOUR CODE HERE...
    if len(list1) != len(list2):
        return "Both lists must have the same length."
    if not (1 <= len(list1) <= 1000) or not ( 1<= len(list2) <= 1000):
        return "The length of the lists must be between 1 and 1000."
    return list(map(lambda x,y: x+y,list1,list2))

list1 = [10, 20, 30]
list2 = [5, 15, 25]
print(add_lists(list1, list2))
```

```python
'''
Write a Python function "add_five_to_divisible_by_three" using "filter" function to filter numbers that are divisible by 3 from a list and add 5 to each of them.

Input:
	* A list of integers numbers where each element is an integer.
	* Use numbers in your implementation.

Output:
	* A list of integers, where each element is a number from numbers that is divisible by 3, with 5 added to it.
	* Result must be "Invalid input: List exceeds maximum length of 1,000 integers." if length of the list is greater than 1000.
	* Result must be "Invalid input: Integers must be between -10,000 and 10,000." if any value in the list is greater than 10000 or less than -10000.

Constraints:
	* Input numbers is list of integers.
'''

def add_five_to_divisible_by_three(numbers):
    # WRITE YOUR CODE HERE...
    if len(numbers)>1000:
        return "Invalid input: List exceeds maximum length of 1,000 integers."
    if any(n<-10000 or n>10000 for n in numbers):
        return "Invalid input: Integers must be between -10,000 and 10,000."
    return list(map(lambda x:x+5,filter(lambda x:x%3==0,numbers)))

numbers = [2, 4, 5, 7]
print(add_five_to_divisible_by_three(numbers))
```

```python
'''
Write a function "is_palindrome(s)" that returns True if the string s is a palindrome (i.e., it reads the same backward as forward), and False otherwise. Then, implement a function "filter_palindromes(strings)" that uses the "filter" function to apply is_palindrome to a dynamically passed list of strings, and returns a list of all palindrome strings from the list.

Input:
	* A list of strings strings, where each string can contain alphabetic characters, numbers, and other symbols.
	* Use strings in your implementation.

Output:
	* A list containing only the palindrome strings from the strings list.
	* Result must be "Invalid Input." if length of the list is greater than 1000 or if length of any value in the list is greater than 100.

Constraints:
	* Input strings is list of datatype string.
'''

def is_palindrome(s):
    # WRITE YOUR CODE HERE...
    return s == s[:: -1]

def filter_palindromes(strings):
    # WRITE YOUR CODE HERE...
    if len(strings) > 1000 or any(len(s) > 100 for s in strings):
        return "Invalid Input."
    palindromes = filter(is_palindrome, strings)
    return list(palindromes)

strings = ["rotator", "pushmeup", "palindrome"]
print(filter_palindromes(strings))
```

```python
'''
You are given a list of strings. Write a function "get_vowel_counts" using "map" function that returns a list containing the number of vowels in each string. A vowel is defined as one of the characters a, e, i, o, u, in either uppercase or lowercase.

Input:
	* A list of strings strings where each element is a string (1 ≤ len(strings) ≤ 1000).
	* Use strings in your implementation.

Output:
	* A list of integers where each element is the count of vowels in the corresponding string from the input list.
	* Result must be "The list must contain between 1 and 1000 strings." if length of the list is not in range 1 and 1000 (inclusive).

Constraints:
	* Input strings is list of datatype string. The elements of the list must either uppercase or lowercase.
'''

def get_vowel_counts(strings):
    # WRITE YOUR CODE HERE...
    if len(strings) < 1 or len(strings) > 1000:
        return "The list must contain between 1 and 1000 strings."
    def ct(s):
        count = 0
        for i in s:
            if i in 'aeiouAEIOU':
                count += 1
        return count
    return list(map(ct, strings))

strings = ["apple", "banana", "cherry"]
print(get_vowel_counts(strings))
```

```python
'''
You are given a list of temperatures in Celsius. Write a function using "map" function "celsius_to_fahrenheit" that converts each Celsius temperature to Fahrenheit and returns a list of the corresponding Fahrenheit temperatures.

Input:
	* A list of integers or floats celsius, where each element represents a temperature in Celsius (1 ≤ len(celsius) ≤ 1000).
	* Use celsius in your implementation.

Output:
	* Result must be "The length of the input list must be between 1 and 1000." if length of the list is not in range 1 and 1000 (inclusive).
	* A list of floats where each element is the temperature in Fahrenheit, converted from the corresponding Celsius temperature.

Constraints:
	* Input celsius is list of integers.
	* The list celsius will contain integers or floats, representing temperatures in Celsius within the range -10**3 to 10**3.

'''

def celsius_to_fahrenheit(celsius):
    # WRITE YOUR CODE HERE...
    if not (1 <= len(celsius) <= 1000):
        return "The length of the input list must be between 1 and 1000."
    return list(map(lambda c: c* 9/5 +32, celsius))
celsius = [0, 100]
print(celsius_to_fahrenheit(celsius))
```
