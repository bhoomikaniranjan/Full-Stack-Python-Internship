```python
'''
Write a Python function calculate_average that takes a list of numbers and a number to skip. The function calculates the average of the numbers in the list, skipping any occurrences of the specified skip number. Use the "pass" statement to handle the skip condition.

Input:
	* A list of integers, numbers_list.
	* An integer, skip_num, which represents the number to skip during calculations.
	* Use numbers_list and skip_num in your implementation.

Output:
	* A float representing the average of the numbers in numbers_list excluding skip_num.
	* If skip_num is not in the numbers_list or list is empty  then print "Empty List".

Constraint:
	* The numeric values present in the numbers_list and skip_num should be in the range (-10**3 to 10**4).
	* Input numbers_list must be a list of numeric values.
	* Input skip_num must be a numeric value
'''

def calculate_average(numbers_list,skip_num):
    # WRITE YOUR CODE HERE...
    if not numbers_list or skip_num not in numbers_list:
        return "Empty List"
    else:
        sum = 0
        count = 0
        for i in numbers_list:
            if i == skip_num:
                pass
            else:
                count += 1
                sum += i
        if sum == 0:
            return "Empty List"
        else:
            return sum/count

numbers_list = [1, 2, 3, 4, 5]
skip_num = 3
print(calculate_average(numbers_list, skip_num))
```

```python
'''
You are given a dictionary scores, where the keys are student names and the values are their corresponding scores. Your task is to write a script that assigns grades based on the scores using the following criteria:

	* If the score is 90 or above, the grade is 'A'.
	* If the score is between 80 (inclusive) and 89, the grade is 'B'.
	* If the score is below 80, the grade is 'C'.
	* Return a dictionary where the keys are the student names, and the values are their corresponding grades.

Input:
	* A dictionary scores where the keys are strings (student names), and the values are integers (scores between 0 and 100).
	* Use scores in your implementation.

Output:
	* A dictionary where the keys are the student names, and the values are their grades based on the given criteria.

Constraints:
	* The dictionary scores will contain student names as strings and their scores as integers between 0 and 100.
	* The number of students will be between 1 and 10^4.
	* Input scores must be dictionary.
'''

def assign_grades(scores):
    # WRITE YOUR CODE HERE...
    grades = {}
    for student,score in scores.items():
        if score >= 90:
            grades[student] = 'A'
        elif score >= 80:
            grades[student] = 'B'
        else:
            grades[student] = 'C'

    return grades
scores = {'Jack': 100, 'Kathy': 85, 'Leo': 79,'David': 92}
print(assign_grades(scores))
```

```python
'''
Write a Python function calculate_numbers that takes three arguments: two integers a and b, and a string operator that specifies the operation to be performed on them. The operation can be one of the following: "add", "sub", "mul", or "div". The function should return the result of the specified operation rounded up to 2 decimal points.

Input:
	* Two numbers a and b (either integers or floating-point numbers).
	* A string operator, which specifies one of the operations: "add", "sub", "mul", or "div".
	* Use a, b and operator in your implementation.

Output:
	* The function should return the result of the specified operation rounded up to 2 decimal points.
	* For "add", return the sum of a and b.
	* For "sub", return the difference between a and b.
	* For "mul", return the product of a and b.
	* For "div", return the quotient of a divided by b.
	        If b = 0, print "Invalid Divisor Value" .
	* Print "Invalid Operator Value" if any other operator defined.

Constraints:
	* Inputs a, b must be numeric values within the range (-10**3 to 10**3)
	* Input operator must be string.
'''

def calculate_numbers(a, b, operator):
    # WRITE YOUR CODE HERE...
    if operator == "add":
        return round(a + b, 2)
    elif operator == "sub":
        return round(a - b, 2)
    elif operator == "mul":
        return round(a * b, 2)
    elif operator == "div":
        if b == 0:
            return "Invalid Divisor Value"
        return round(a / b, 2)
    else:
        return "Invalid Operator Value"

a = 4
b = 0
operator = "div"
print(calculate_numbers(a, b, operator))
```

```python
'''
Write a recursive Python function factorial(n) that computes the factorial of a non-negative integer n. The function should return 1 for n = 0.

Input
	* n is a non-negative integer from 1 to 1000 (inclusive).
	* Use n in your implementation.

Output:
	* The function should return the factorial of n, defined as:
	           n! = n * (n - 1) * (n - 2) * ... * 1
	           Special case: 0! = 1.
        * n will be a non-negative integer in the range [0, 1000] (inclusive). Return "Invalid Value" if a negative value or a value above 1000 is entered.

Constraints:
	* Input n must be a integer.
'''

def factorial(n):
    # WRITE YOUR CODE HERE...
    if n < 0 or n > 1000:
        return "Invalid Value"
    if n == 0:
        return 1
    return n * factorial (n - 1)

n = 10
print(factorial(n))
```

```python
'''
Write a Python function calculate_area_and_circumference(radius) that calculates the area and circumference of a circle given its radius. The function should return these two values as a tuple (area, circumference).
Note: Use the value of pi as 3.14.

Input:
	* A single argument: radius (float or integer), which is the radius of the circle.
	* Use radius in your implementation.

Output:
	* A tuple (area, circumference) where
	* The area of the circle is calculated as π × radius^2
 	* The circumference of the circle is calculated as 2 × π × radius
	* Print "Invalid Radius" if radius is less than 0.

Constraints:
	* The arguments radius can be of any numeric type within the range -10**3 to 10**3.
	* Input radius must be a numeric value.
'''

def calculate_area_and_circumference(radius):
    # WRITE YOUR CODE HERE...
    if radius < 0:
        return "Invalid Radius"

    pi = 3.14
    area = pi * radius**2
    circumference = 2 * pi * radius
    return (area, circumference)

radius = 6.0
print(calculate_area_and_circumference(radius))
```

```python
'''
Question:
You are given a dictionary original_dict where the keys are strings, and the values are integers. Your task is to create a new dictionary that includes only the key-value pairs where the key starts with a specific letter. You should return the new dictionary, which includes only the entries where the key begins with that specified letter.

Input:
	* A dictionary original_dict where the keys are strings and the values are integers.
	* A string letter which is a single lowercase letter.
	* Use original_dict in your implementation.

Output:
	* A new dictionary that includes only the entries where the key starts with the specified letter.

Constraints:
	* The dictionary original_dict contains keys of type string and values of type integer.
	* The length of original_dict will be between 1 and 10^4 key-value pairs.
	* Input original_dict must be a datatype of dictionary
	* The letter will always be a single lowercase letter.

'''

def filter_dict_by_letter(original_dict, letter):
    # WRITE YOUR CODE HERE...
    return {key: value for key, value in original_dict.items() if key.startswith(letter)}

original_dict = {'apple': 1, 'banana': 2, 'apricot': 3}
letter = ''
print(filter_dict_by_letter(original_dict, letter))
```
