```python
'''
Write a Python function using a "break" statement that takes a string as input and iterates through its characters to find the first vowel. The loop should exit as soon as a vowel is encountered, using a break statement. Print the first vowel if it exists, or print "No vowel found" if there are none.

Input:
	* A single string  containing uppercase and lowercase English letters.
	* The string must be used in your implementation.

Output:
	* If a vowel is found, print: "First vowel is '<vowel>'".
	* If no vowel is found, print: "No vowel found".
	* If the input list is not between 1 and 100 (inclusive), print an error message: "Error: The string must contain between 1 and 100 characters".

Constraints:
	* The input string must be of datatype string.
'''

def find_first_vowel(string):
    # WRITE YOUR CODE HERE...
    if len(string) < 1 or len(string) > 100:
        print("Error: The string must contain between 1 and 100 characters")
        return
    vowels = "aeiouAEIOU"
    for char in string:
        if char in vowels:
            print(f"First vowel is:'{char}'")
            break
    else:
        print("No vowel found")

string = "a"*101
find_first_vowel(string)
```

```python
'''
Write a python script that prints the ASCII values of each character in a given string using a "while" loop. For each character in the string, the script should print the character along with its ASCII value on a new line.

Input:
	* A string containing any printable ASCII characters. The string can include letters (both uppercase and lowercase), digits, punctuation, and whitespace.
	* Use variable string in your implementation.

Output:
	* The script should print the ASCII value of each character in the string
	* Print "Invalid String" if the input is empty.

Constraints:
	* The input string should be of datatype string.
'''

def ascii_values(string):
    if not string:
        print("Invalid String")
        return

    index = 0
    while index < len(string):
        char = string[index]
        print(f"ASCII value of '{char}' is {ord(char)}")
        index += 1

string = "1234"
ascii_values(string)
```

```python
'''
Write a Python function using "break" statement that finds the first word in a given string that contains more than 5 letters. If such a word is found, print the word. If no word has more than 5 letters, print "No word with more than 5 letters found"

Input:
	* The input string can contain between 1 and 100 words, separated by spaces.
	* Use words in your implementation.

Output:
	* If a word with more than 5 letters is found, print the word.
	* If no word has more than 5 letters, print "No word with more than 5 letters found"
	* For a word count outside the 1 to 100 range (inclusive), print: "Error: The input must contain between 1 and 100 words".

Constraints:
	* The words must be of datatype string of words.
'''

def find_first_long_word(words):
    # WRITE YOUR CODE HERE
    words_list = words.split()
    if len(words_list) < 1 or len(words_list) > 100:
        print("Error: The input must contain between 1 and 100 words")
        return
    for word in words_list:
        if len(word) > 5:
            print(word)
            break
    else:
        print("No word with more than 5 letters found")

words = "treehouse playground sunflower guitar bicycle"
find_first_long_word(words)
```

```python
'''
Write a Python program that prints the days of the week but skips weekends (Saturday and Sunday). Use the "continue" keyword to bypass weekends.

Input:
	* A list of strings representing the days of the week.
	* The days must be used in your implementation.

Output:
	* The program should print each weekday (Monday to Friday) on a new line, skipping Saturday and Sunday entered in any case either upper or lower.
	* Print "The list is empty." if the list is empty.

Constraints:
	* The input days must be list of days in a week.

'''

def days_of_week(days):
    # WRITE YOUR CODE HERE...
    if not days:
        print("The list is empty.")
        return

    for day in days:
        if day.lower() in ["saturday","sunday"]:
            continue
        print(day)

days = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
days_of_week(days)
```

```python
'''
Write a Python program that removes all vowels from a string and prints the result. Use the "continue" keyword to skip vowels.

Input:
	* A single string text (1 ≤ length of text ≤ 100), which can contain both upper and lower case letters, spaces, punctuation marks, and other special characters.
	* Use text in your implementation.

Output:
	* The modified string with all vowels (a, e, i, o, u in both uppercase and lowercase) removed.
	* "Input string length must be between 1 and 100 characters." if length of string is less than 0 or greater than 100.

Constraints:
	* The text must be a string.
'''

def remove_vowels(text):
    # WRITE YOUR CODE HERE...
    if len(text) < 1 or len(text) > 100:
        return "Input string length must be between 1 and 100 characters."
    vowels = "aeiouAEIOU"
    result = ""
    for char in text:
        if char in vowels:
            continue
        result += char
    return result

text = ""
print(remove_vowels(text))
```

```python
'''
Write a Python program using "break" statement that classifies a list of ages into categories: "Child" (0-12), "Teenager" (13-19), "Adult" (20-64), or "Senior" (65+). The classification process should stop when the first negative number is encountered in the list, and then the program should print the classifications for all valid ages entered before the negative number.

Input:
	* A list named ages_list containing various age values (both valid ages and possibly negative numbers).
	* Use ages_list in your implementation.

Output:
	* Print the classification for each valid age before the first negative number.

Constraints:
	* The ages_list must be list of integers with length of list greater than 0.
'''

def age_classification(ages_list):
    # WRITE YOUR CODE HERE...
    if len(ages_list) == 0:
        print("Error: The list must contain at least one age")
        return
    for age in ages_list:
        if age < 0:
            break
        if age <= 12:
            print("Child")
        elif age <= 19:
            print("Teenager")
        elif age <= 64:
            print("Adult")
        else:
            print("Senior")

ages_list = [50,87,45,32]
age_classification(ages_list)
```

```python
'''
Write a Python program using "break" statement that checks a predefined password against a user-provided password. Allow up to 3 attempts to check the correct password (secure123) in the list. If the correct password is entered within 3 attempts, print "Access granted." Otherwise, print "Access denied."

Input:
	* The user will input their password, and the program will validate it against the predefined password.
	* Use entries in your implementation.

Output:
	* Print "Access granted." if the correct password is found within 3 elements in the list.
	* Print "Access denied." if the correct password is not found within 3 elements in the list.
	* Print "Wrong password" if the element in the list does not match with the correct password.
	* Print "No passwords entered." if the list is empty.

Constraints:
	* The entries must be list of datatype string.
'''

def password_access(entries):
    # WRITE YOUR CODE HERE...
    if not entries:
        print("No passwords entered.")
    else:
        count = 0
        for i in entries:
            if i == 'secure123':
                print("Access granted.")
                break
            else:
                print("Wrong password")
                count += 1
            if count == 3:
                print("Access denied.")
                break

entries = ['password123', '12345', 'secure123', 'qwerty', 'letmein']
password_access(entries)
```

```python
'''
Write a Python program using "break" statement that calculates the sum of all positive numbers in a given list. The program should stop processing once a negative number is encountered and then print the total sum of the entered positive numbers.

Input:
	* The input list can contain positive numbers, zero, and negative numbers. The program should stop summing when the first negative number is found.
	* Use numbers_list in your implementation.

Output:
	* Print the total sum of the positive numbers encountered before the first negative number.

Constraints:
	* The numbers_list must be list of integers.
'''

def calculate_sum(numbers_list):
    # WRITE YOUR CODE HERE...
    total_sum = 0
    for number in numbers_list:
        if number < 0:
            break
        total_sum += number
    print(f"Total sum of positive numbers: {total_sum}")

numbers_list = [10,20,30,-10,20]
calculate_sum(numbers_list)
```

```python
'''
Write a Python function that iterates through a given list of words and prints only those words that do not start with the letter 'A' or 'a'. Use a "continue" statement to skip any words starting with 'A' or 'a'.

Input:
	* A single list words containing strings of words separated by commas.
	* Use words in your implementation.

Output:
	* Print each word that does not start with the letter 'A' or 'a' on a new line.
	* Print "Error: List must contain between 1 and 100 words, each between 1 and 20 characters." if length of list not between 1 and 100 (inclusive) or length of each word is not between 1 and 20 (inclusive).
	* Print "No words processed" if all the words in the list start with letter "a".

Constraints:
	* Input words must be list of strings.
'''

def print_non_a_words(words):
    # WRITE YOUR CODE HERE...
    if not (1 <= len(words) <= 100) or any (not (1 <= len(word) <=20) for word in words):
        print("Error: List must contain between 1 and 100 words, each between 1 and 20 characters.")

        return
    processed = False
    for word in words:
        if word[0].lower() == 'a':
            continue
        print(word)
        processed = True
    if not processed:
        print("No words processed")


words = ["apple", "banana", "apricot", "cherry", "Avocado", "fig"]
print_non_a_words(words)
```

```python
'''
Write a Python program that processes a list of exam scores, but skips any scores below 50 or above 100 using "continue". Print the valid scores that are in between 50 and 100 (inclusive).

Input:
	* The input will be a list of integers representing the exam scores.
	* Use scores in your implementation.

Output:
	* The output will be the valid scores that are in between 50 and 100 (inclusive), printed one per line.
	* If the list is empty print "No scores to process.".

Constraints:
	* Input scores must be list of integers
'''

def valid_score(scores):
    # WRITE YOUR CODE HERE...
    if not scores:
        print("No scores to process.")
        return
    for score in scores:
        if score < 50 or score > 100:
            continue
        print(f"valid score: {score}")

scores = [45, 55, 60, 30, 70, 100, 50]
valid_score(scores)
```

```python
'''
Write a Python program that reads a list of integers and prints only the positive numbers. Use the "continue" keyword to skip non-positive numbers.

Input:
        * A list of integers where list length is in range 1 and 100 (inclusive). Each integer in the list is between -1000 and 1000 (inclusive).
	* The numbers must be used in your implementation.

Output:
        * Print each positive integer from the list on a new line, in the order they appear in the input list.
	* If the input list is not between 1 and 100 (inclusive) and numbers in the list is not between -1000 and 1000 (inclusive), print an error message: "Input list does not meet the constraints.".

Constraints:
	* The input numbers must be list of numbers.
'''

def get_positive_numbers(numbers):
    # WRITE YOUR CODE HERE...
    if not (1 <= len(numbers) <= 100) or any(not (-1000 <= num <= 1000) for num in numbers):
        print("Input list does not meet the constraints.")
        return

    for num in numbers:
        if num <= 0:
            continue
        print(num)

numbers = [1001, 20, -30]
get_positive_numbers(numbers)
```

```python
'''
Write a Python program that calculates the sum of positive numbers in a list. Use the "continue" keyword to skip negative numbers.

Input:
	* The input list can contain positive numbers, zero, and negative numbers.
        * A list of integers where list length is in range 1 and 100 (inclusive). Each integer in the list is between -1000 and 1000 (inclusive).
	* Use numbers in your implementation.

Output:
	* Print the total sum of the positive numbers.
	* If the input list is not between 1 and 100 (inclusive) and numbers in the list is not between -1000 and 1000 (inclusive), print an error message: "Input list does not meet the constraints.".

Constraints:
	* The numbers_list must be list of integers.
'''

def sum_of_positive_numbers(numbers):
    # WRITE YOUR CODE HERE...
    if not (1 <= len(numbers) <= 100) or any (not(-1000 <= num <= 1000) for num in numbers):
        return "Input list does not meet the constraints."
    total_sum = 0
    for num in numbers:
        if num <= 0:
            continue
        total_sum += num
    return f"Sum of positive numbers: {total_sum}"

numbers = [10, -5, 3, -1, 7, -2]
print(sum_of_positive_numbers(numbers))
```

```python
'''
Write a Python program using "break" statement that checks a list of email addresses and identifies which ones are valid. An email address is considered valid if it contains an "@" symbol, has some characters between "@" and ".", includes a "." symbol after the "@" and has characters after the ".". The program should stop checking the list once it finds the first valid email address and print it. If no valid email addresses are found, indicate that no valid emails are present.

Input:
	* A list named email_list containing multiple email addresses to validate.
	* You must use email_list in your implementation.

Output:
	* Print the first valid email address found in the list, or indicate.
	* Print "No valid email addresses found." if no email is found.

Constraints:
	* Input email_list must be list of strings.
'''

def valid_email(email_list):
    # WRITE YOUR CODE HERE...
    for email in email_list:
        if "@" in email and "." in email:
            at_index = email.find("@")
            dot_index = email.find(".",at_index)
            if at_index > 0 and dot_index > at_index + 1:
                print(f"First valid email address found: {email}")
                break
    else:
        print("No valid email addresses found.")


email_list = ["plainaddress", "@missingusername.com", "username@.com"]
valid_email(email_list)
```
