```python
'''
Write a Python function to classify a person's age into one of the following categories:
	* "Child" for ages 0 to 12.
	* "Teenager" for ages 13 to 19.
	* "Adult" for ages 20 to 59.
	* "Senior" for ages 60 and above.

Input:
	* An integer age representing a person's age.
	* You must use the variable age in your implementation.

Output:
	* A string indicating the classification:
	* "Child" if age is between 0 and 12 (inclusive).
	* "Teenager" if age is between 13 and 19 (inclusive).
	* "Adult" if age is between 20 and 59 (inclusive).
	* "Senior" if age is 60 or more.
	* "Error: Age must be between 0 and 120." if age is not in range 0 to 120 (exclusive).

Constraints:
	* Input age should be an integer.
'''

def classify_age(age):
    if age < 0 or age >120:
        print("Error: Age must be between 0 to 120.")
    elif 0 <= age <= 12:
        print("Child")
    elif 13 <= age <= 19:
        print("Teenager")
    elif 20 <= age <= 59:
        print("Adult")
    elif age >= 60:
        print("Senior")

age = int(input())
classify_age(age)
```

```python
'''
Write a Python function to calculate the bonus based on years of service:
	* "10% Bonus" if years of service are 10 or more.
	* "5% Bonus" if years of service are between 5 and 9.
	* "No Bonus" if years of service are less than 5.

Input:
	* An integer years representing the number of years of service.
	* You must use the variable years in your implementation.

Output:
	* A string indicating the bonus:
	* "10% Bonus" if years is 10 or more.
	* "5% Bonus" if years is between 5 and 9 (inclusive).
	* "No Bonus" if years is less than 5.
	* "Error: Years of service must be between 0 and 50." if years is not in range 0 to 50 (exclusive).

Constraints:
	* Input years should be integer.
'''

def calculate_bonus(years):
    if years < 0 or years > 50:
        print("Error: Years of service must be between 0 and 50.")
    else:
        if years >= 10:
            print ("10% Bonus")
        elif years >= 5:
            print ("5% Bonus")
        else:
            print ("No Bonus")
print(calculate_bonus(102))
```

```python
'''
You are tasked with writing a Python function that calculates the sum of all given numbers in a list using a for loop.

Input:
	* A list of integers or floats numbers that contains the numbers to be summed.
	* Use input number in your implementation.

Output:
	* Print the sum of all the numbers in the list.

Constraints:
	* The number must be a list of integer.
'''

def get_sum_of_list(numbers):
    # WRITE YOUR CODE HERE...
    if not isinstance(numbers,list) or not (isinstance(num,(int,float)) for num in numbers):
        print("Invalid input")
        return
    total = 0
    for num in numbers:
        total += num
    print(total)
numbers = [1,2,3,4,5]
get_sum_of_list(numbers)
```

```python
'''
Write a Python script that checks if a given character is a vowel or a consonant using a ternary conditional expression. The script should print "Vowel" if the character is a vowel (a, e, i, o, u) and "Consonant" if it is a consonant.

Input:
	* A single character char (a string of length 1).
	* char can be any lowercase or uppercase English alphabet letter (a to z).
	* Use input char in your implementation.

Output:
	* Print "Vowel" if the character is a vowel.
	* Print "Consonant" if the character is a consonant.
	* Print "Invalid Char" if the above constraint fails.

Constraints:
	* Input char should be a string.
'''

def check_vowel_or_consonant(char):
    # WRITE YOUR CODE HERE...
    if len(char) == 1 and char.isalpha():
        print("Vowel" if char.lower() in 'aeiou' else "Consonant")
    else:
        print("Invalid Char")
char = 'b'
check_vowel_or_consonant(char)
```

```python
'''
Write a function check_empty_string that checks if a given string is empty or not. The function should return "Empty" if the string is empty, and "Not Empty" if the string contains any characters.

Input:
	* A string string (0 <= len(string) <= 1000), which may contain any printable ASCII characters (including spaces, letters, digits, and symbols).
	* Use input string in your implementation.

Output:
	* The function should return a string.
	* "Empty" if the input string is empty.
	* "Not Empty" if the input string is not empty (contains at least one character).
	* Print "Invalid String" if the length of string is greater than 1000.

Constraints:
	* The input string must be of datatype string.
'''

def check_empty_string(string):
    # WRITE YOUR CODE HERE...
    if not isinstance(string, str):
        print("Invalid String")
    elif len(string) > 1000:
        print("Invalid String")
    elif len(string) == 0:
        print("Empty")
    else:
        print("Not Empty")
string = "how are you"
check_empty_string(string)
```

```python
'''
Write a Python function to determine the discount percentage based on a given price.

Input:
	* An integer value price representing the price of an item.
	* You must use the variable price in your implementation.

Output:
	* A string indicating the discount percentage:
	* "20% discount" if price is 1000 or more.
	* "10% discount" if price is between 500 (inclusive) and 1000 (exclusive).
	* "5% discount" if price is less than 500.
        * "Error: Price must be between 1 and 100,000." if price is less than or equal to 0 and price greater than 100,000.

Constraints:
	* Input price should be an integer.
'''

def discount(price):
    if price <= 0 or price > 100000:
        print("Error: Price must be between 1 and 100,000.")
    elif price >= 1000:
        print("20% discount")
    elif 500 <= price <1000:
        print("10% discount")
    else:
        print("5% discount")

price = 1200
discount(price)
```

```python
'''
You are tasked with writing a Python program that evaluates a student's grade based on their score using ternary operators.

Input:
	* An integer score representing the student's score (0 to 100).
	* Use input score in your implementation.

Output:
	* Return "A" if the score is 90 or above.
	* Return "B" if the score is between 80 and 89 (inclusive).
	* Return "C" if the score is between 70 and 79 (inclusive).
	* Return "D" for any score below 70.
	* Return "Invalid score." for any score exclusive between the range 0 to 100.

Constraints:
	* The score must be an integer.
'''

def get_grade(score):
    # WRITE YOUR CODE HERE...
    if score < 0 or score > 100:
        print("Invalid score.")
    else:
        grade = (
            "A" if score >= 90 else "B" if score >= 80 else "C" if score >= 70 else "D")
        print(grade)
score = 0
get_grade(score)
```

```python
'''
You are tasked with writing a Python program that prints all even numbers from 1 to a given number using a for loop.

Input:
	* An integer number representing the upper limit (inclusive) of the range. The minimum value for number must be 2.
	* Use input number in your implementation.

Output:
	* Print all even numbers between 1 and the given number (inclusive).
	* Print "Invalid input. The number must be 2 or greater." if number is less than 2.

Constraints:
	* The number must be an integer.
'''

def get_even_numbers(number):
    # WRITE YOUR CODE HERE...
    if number < 2:
        print("Invalid input. The number must be 2 or greater.")
    else:
        for num in range (1, number +1):
            if num % 2 == 0:
                print(num)
number = 2
get_even_numbers(number)
```

```python
'''
Write a function check_three_digit_number that checks if a given number is a three-digit number or not. The function should print the string "Three-Digit Number" if the number is a three-digit number, and "Not a Three-Digit Number" if it is not. A three-digit number is any number that has exactly three digits, i.e., in the range 100 to 999 (inclusive), or -100 to -999 (inclusive) for negative numbers.

Input:
	* A single integer num (-10000 <= num <= 10000), which represents the number to check.
	* Use input num in your implementation.

Output:
	* The function should print a string.
	* Three-Digit Number" if the number is a three-digit number.
	* "Not a Three-Digit Number" if the number is not a three-digit number.
	* Print "Invalid Number" if num not in range -10000 to 10000, inclusive.

Constraints:
	* Input num must be an integer.
'''

def check_three_digit_number(num):
    # WRITE YOUR CODE HERE...
    if num < -10000 or num > 10000:
        print("Invalid Number")
    elif 100 <= abs(num) <= 999:
        print("Three-Digit Number")
    else:
        print("Not a Three-Digit Number")
num = 1234
check_three_digit_number(num)
```

```python
'''
You are tasked with writing a Python program that checks whether a given year is a leap year.

Input:
	* An integer year representing the year to be checked.
        * You must use the variable "year" in your implementation.

Output:
	* Print "Leap year" if the year is a leap year.
	* Print "Not a leap year" if the year is not a leap year.
	* The year must be a positive integer (greater than 0). Print "Invalid year" if the constraints fails.
	* A year is a leap year if:
              It is divisible by 4.
	      It is not divisible by 100, unless it is also divisible by 400.

Constraints:
	* Input year must be an integer.
'''

def check_leap(year):
    if year <= 0:
        print("Invalid year")
    elif year % 4 == 0:
        if year % 100 == 0:
            if year % 400 == 0:
                print("Leap Year")
            else:
                print("Not a leap year")
        else:
            print("Leap Year")
    else:
        print("Not a leap year")

year = int(input())
check_leap(year)
```
